### Importing modules and data

```python
import pandas as pd
import matplotlib.pyplot as plt
import contextily
# Read the restaurants csv file
restaurants = pd.read_csv("paris_restaurants.csv")
restaurants.head()
```

## Regular Scatterplots

```python
plt.plot(restaurants.x,restaurants.y,'o')
plt.title('Regular geospatial coordinates plot')
plt.show()
```

## Plots with contextily

```python
fig,ax = plt.subplots()
ax.plot(restaurants.x,restaurants.y,'o',markersize=1)
contextily.add_basemap(ax)
plt.title('Base map background')
plt.show()
```
