## Importing the modules and the data

```python
import pandas as pd
import matplotlib.pyplot as plt
import contextily
restaurants = pd.read_csv('paris_restaurants.csv')
restaurants.head()
```

## Regular Plots

```python
fig,ax = plt.subplots()
ax.plot(restaurants.x,restaurants.y,'o')
plt.title('Regular Scatter plot with geospatial coordinates')
plt.show()
```

## Map Plots

```python
fig,ax = plt.subplots()
ax.plot(restaurants.x,restaurants.y,'o', markersize = 1)
contextily.add_basemap(ax)
plt.title('Scatter plot with geospatial coordinates on a map background')
plt.show()
```
