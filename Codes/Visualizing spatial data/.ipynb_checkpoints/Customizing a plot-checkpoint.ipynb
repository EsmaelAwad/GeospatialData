```python
import geopandas as gpd
districts = gpd.read_file('paris_districts_utm.geojson')
districts.head()
```

## Inspecting the area of districts

```python
districts.area # The area is squared meter, for example : the first polygon is 8.6 meters squared.
```

```python
# Add a population density column (number of people per squared kilometer)
districts['Population Density'] = districts.population / districts.area * 10**6
districts.head()
```

## Plot of districts colored by the population density

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10,10))
districts.plot(ax = ax , column = 'Population Density' , legend = True)
ax.set_axis_off() # To hide the axis
plt.show()
```

```python
import pandas as pd
restaurants_regular = pd.read_csv('paris_restaurants.csv')
restaurants_geo = gpd.GeoDataFrame(restaurants_regular,geometry = gpd.points_from_xy\
                                   (restaurants_regular.x,restaurants_regular.y))
restaurants_geo.drop( ['x','y'], axis = 1, inplace=True)
restaurants_geo.head()
```

```python
restaurants_geo['type'].value_counts()
#You can get the same result using: restaurants.groupby('type').size()
```

## Where exactly are the african restaurants?

```python
african_rest = restaurants_geo[restaurants_geo['type'] == 'African restaurant']
fig , ax = plt.subplots(figsize = (10,10))
restaurants_geo.plot(color = 'grey',ax = ax)
african_rest.plot(ax = ax , color = 'red')
ax.set_axis_off() # To hide the axis
plt.title("As you can see, Most of em are in the north of the center")
plt.show()
```
